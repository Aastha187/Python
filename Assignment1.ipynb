## Part 2: Basic python syntax

```python
firstName=input("Enter Your First Name: ") #Taking First name as input
lastName=input("Enter Your Last Name: ") #Taking last name as input
fullName=firstName+" "+lastName #Merging First and last name 
reversedName = fullName[::-1] #reverse full name using slicing
print(reversedName) #Printing Reversed name
```

```python
str1="hello aastha!"
print("Method 1: ",str1.capitalize())#Convert First character of first word to uppercase
str2="Hello Aastha!"
print("Method 2: ",str2.swapcase())#Swap the case of all character it converts lowercase to uppercase and uppercase to lowercase
```

```python
num=input("Enter number: ") #Taking input from user
intnum=int(num) #Integer (int): Represents whole numbers without decimals (e.g., 5, -3).
floatnum=float(num) #Float (float): Represents real numbers with decimal points (e.g., 3.14, -0.001).
compnum=complex(num) #Complex (complex): Represents numbers with a real and an imaginary part (e.g., 2 + 3j, -1j).
print("\nInteger Form: ", intnum)
print("Float Form: ", floatnum)
print("Complex Form: ", compnum)
```

```python
# Taking length and breadth as input
length=int(input("Enter Length: "))
breadth=int(input("Enter Breadth: "))
# Creatind function to calculate area
def arearec(l,b):
    area=l*b
    print("Area: ",area)
arearec(length,breadth)  # Calling functiion
```

```python
def arearec(l,b):
    area=l*b
    print("Area: {:.2f}".format(area))  #Using slicing and format method to print 
arearec(length,breadth)
```

```python
#Taking inputs
num1=int(input("Enter First Number: "))  
num2=int(input("Enter Second Number: "))
num3=int(input("Enter Third Number: "))
average=(num1+num2+num3)/3  #calculating average
print("The average of the three numbers is: %d"%(average)) #Printing average using % method
print("The average of the three numbers is: ",int(average)) #Printing average using print function
```

## Part 3: Language Components

```python
while True:
    number=input("Enter Number: ")
    if number.lower() == 'exit':
        print("Exited!")
        break
    # Try to convert the input to a float
    try:
        number = int(number)
        # Determine if the number is positive, negative, or zero
        if number > 0:
            print("Number is positive.")
        elif number < 0:
            print("Number is negative.")
        else:
            print("Number is zero.")  
    except ValueError:
        print("Invalid input!")
```

```python
# Take input from the user
number1 = int(input("Enter the first number: "))
number2 = int(input("Enter the second number: "))

# Check if both numbers are even
if number1 % 2 == 0 and number2 % 2 == 0:
    print("Both numbers are even.")
# Check if both numbers are odd
elif number1 % 2 != 0 and number2 % 2 != 0:
    print("Both numbers are odd.")
else:
    print("One number is even and one is odd.")
```

```python
def convert_number(num):
    print(f"Decimal: {num}")
    
    # Binary
    print("Binary: ", end="")
    for i in range(31, -1, -1):
        print(1 if num & (1 << i) else 0, end="")
    print()
    
    # Octal
    print("Octal: {:o}".format(num))
    
    # Hexadecimal
    print("Hexadecimal: {:X}".format(num))

# Get user input
decimal_num = int(input("Enter a decimal number: "))

# Call the function to convert the number
convert_number(decimal_num)
```
